```python

from bs4 import BeautifulSoup
import pandas as pd
import requests
page = requests.get('https://somethinc.com/en/collection/makeup/')
bs = BeautifulSoup(page.text, 'html.parser')
bs

# Buat list kosong untuk menampung hasil scraping
products = []

for page in range(1, 3):

    # Iterasi setiap elemen dengan class 'caption'
    for item in bs.find_all('div', class_='caption'):
        title = item.find('div', class_='title').text.strip()
        
        # Cek apakah old_price ada atau tidak
        old_price = item.find('s')
        old_price = old_price.text.strip() if old_price else 'n/a'
        
        current_price = item.find('span', class_='text-danger')
        current_price = current_price.text.strip() if current_price else old_price

        # Bersihkan string harga dengan menghapus 'Rp ' dan titik
        old_price_clean = old_price.replace('Rp ', '').replace('.', '') if old_price != 'n/a' else 'n/a'
        current_price_clean = current_price.replace('Rp ', '').replace('.', '') if current_price != 'n/a' else 'n/a'

        # Hitung diskon jika ada
        if old_price_clean != 'n/a' and current_price_clean != old_price_clean:
            discount = round((float(old_price_clean) - float(current_price_clean)) / float(old_price_clean) * 100)
        else:
            discount = 'n/a'
        
        # Masukkan hasil scraping ke list products
        products.append({
            'Product': title,
            'Old Price': old_price,
            'Current Price': current_price,
            'Discount (%)': f'{discount}%' if discount != 'n/a' else 'n/a'
        })

# Buat DataFrame dari list products
df = pd.DataFrame(products)
df.head(60)
# # Tampilkan tabel
# print(df)
```
